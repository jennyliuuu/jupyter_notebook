```python
# https://blog.techbridge.cc/2017/07/28/data-science-101-numpy-tutorial/

# 引入 numpy 模組
import numpy as np
np1 = np.array([1, 2, 3])
np2 = np.array([3, 4, 5])
# 陣列相加
print("np1 + np2: ", np1 + np2) # [4 6 8]
# 顯示相關資訊
print("\n", np1.ndim, np1.shape, np1.dtype) # 1 (3,) int64 => 一維陣列, 三個元素, 資料型別

np3 = np.array([1, 2, 3, 4, 5, 6])
print("\n np3: ", np3)
np3 = np3.reshape([2, 3])
print("\n np3.reshape([2, 3]): ", np3)

# array with 0 or 1
np4 = np.zeros([2, 3]) # array([[ 0.,  0.,  0.], [ 0.,  0.,  0.]])
np5 = np.ones([2, 3]) # array([[ 1.,  1.,  1.], [ 1.,  1.,  1.]])
print("\nnp4: ", np4) 
print("\nnp5: ", np5)

np3 = np.array([1, 2, 3, 4, 5, 6])
print("\n np3 > 3: ", np3 > 3) # [False False False  True  True  True]
print("\n np3[np3 > 3]: ", np3[np3 > 3]) # [4 5 6]
print("\n np3[np3 > 3][0]: ", np3[np3 > 3][0])
print("\n np3[np3 > 3][1]: ", np3[np3 > 3][1])
print("\n np3[np3 > 3][2]: ", np3[np3 > 3][2])
np3 = np3.reshape([2, 3])
print("\n", np3.sum(axis=1)) # 將 axis=1 column加總 [6 15]
# axis=0 row, axis=1 column
```

## [資料分析&機器學習] 第2.4講：資料前處理(Missing data, One-hot encoding, Feature Scaling)

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from io import StringIO
```

```python
csv_data = '''A,B,C,D,E
            5.0,2.0,3.0,,6
            1.0,6.0,,8.0,5
            0.0,11.0,12.0,4.0,5
            3.0,,3.0,5.0,
            5.0,1.0,4.0,2.0,4
           '''
```

```python
df = pd.read_csv(StringIO(csv_data))
df
```

### Missing Data

```python
df.dropna()
```

```python
df.dropna(how='all')  # 把how設為all，就是要全部為空才清掉
```

```python
df.dropna(subset=['C'])
```

```python
df.fillna(0)
```

```python
df['B'] = df['B'].fillna(df['B'].mean())
df
```

```python
df['C'] = df['C'].fillna(df['C'].mean())
df
```

```python
df['D'] = df['D'].fillna(df['D'].median())
df
```

```python
df['E'] = df['E'].fillna(df['E'].min())
df
```

### Categorical Data

```python
df2 = pd.DataFrame(
    [['green', 'M', 10.1, 1],
    ['red', 'L', 13.5, 2],
    ['blue', 'XL', 15.3, 1]]
)
df2.columns = ['color', 'size', 'price', 'classlabel']
print("\n", df2)

### the same as ###

df3 = pd.DataFrame(
    [['green', 'M', 10.1, 1],
    ['red', 'L', 13.5, 2],
    ['blue', 'XL', 15.3, 1]], columns=['color', 'size', 'price', 'classlabel']
)
print("\n", df3)
```

```python
size_mapping = {
    'XL':3,
    'L':2,
    'M':1
}
df2['size'] = df2['size'].map(size_mapping)
df2
```

```python
pd.get_dummies(df2['color'])  # 在pandas裡面要使用onehot-encoding使用get_dummies這個函式就可以了
```

```python
onehot_encoding = pd.get_dummies(df2['color'], prefix = 'color')
onehot_encoding
```

```python
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.drop.html
df2 = df2.drop(labels='color', axis=1)  
df2
```

```python
pd.concat([onehot_encoding, df2],axis=1)
```

### 資料正規化(normalization)

```python
from IPython.display import Math
```

```python
iris = datasets.load_iris()
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
print("target_names: "+str(iris['target_names']))
y = pd.DataFrame(iris['target'], columns=['target_names'])
data = pd.concat([x,y], axis=1)
data.head(3)
```

```python
data['sepal length (cm)'] = (data['sepal length (cm)'] - data['sepal length (cm)'].min())/\
                            (data['sepal length (cm)'].max() - data['sepal length (cm)'].min())
data.head()
```

### 資料標準化(Standardization)

```python
data['sepal width (cm)'] = (data['sepal width (cm)'] - data['sepal width (cm)'].mean())/\
                            (data['sepal width (cm)'].std())
data.head()
```
